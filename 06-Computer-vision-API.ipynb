# Custom Vision AI APP with Azure

1. Create a new Custom Vision AI resource in Azure. Choose the Free pricing tier.
2. Now let's go to the training new resource.
3. Create a new project. Choose "Classification" and "Multiclass (single tag per image)".
4. Upload images for each class you want to classify. Lets use those images (apples, bananas and oranges). https://github.com/MicrosoftLearning/mslearn-ai-vision/tree/main/Labfiles/image-classification/training-images 
5. Make sure to tag them correctly.
6. Once you have uploaded enough images for each class, click on "Train" to train your model (choose "quick training").
7. Now use the "Quick Test" feature to test your model with some sample images. You can pass directyly image URLs from the web.
